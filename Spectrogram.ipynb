```python
import matplotlib.pyplot as plt
import numpy as np
x=np.loadtxt('vela_Pulsar.mbr')
x[:, 1]
```

```python
y=np.reshape(x[:,1],(512,60000),order = 'F')
v2=np.zeros((512,60000))
for i in range(60000):
    v2[:,i]=np.power(abs(np.fft.fft(y[:,i])),2)
```

```python
x5=np.zeros((512,1000))
for i in range(1000):
    x5[:,i]=np.sum(v2[:,60*i:60*(i+1)-1],axis=1)
```

```python
plt.pcolormesh(x5[0:255,:])
plt.xlabel('Time')
plt.ylabel('Frequency')
plt.colorbar()
plt.show()
```

```python
yx=np.reshape(x[:,1],(1,30720000))
xfft=np.zeros((1,30720000))
for i in range(30720000):
    xfft[:,i]=np.power(abs(np.fft.fft(yx[:,i])),2)
```
