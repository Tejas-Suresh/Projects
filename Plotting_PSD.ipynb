```python
import matplotlib.pyplot as plt
import numpy as np
import h5py as hp
x=np.array(2)
#print(.getcwd()) 
hp=hp.File('GData.hdf5', 'r')
hp.keys()
dset = hp['strain']["Strain"].value
dset.shape
#plt.hist(v2,512)
#plt.show()
#x[:, 1]
```

```python
plt.plot(dset)
plt.xlabel('GPS Time')
plt.ylabel('Strain')
plt.show()
```

```python
y=np.reshape(dset,(512,32768),order = 'F')
v2=np.zeros((512,32768))
for i in range(32768):
    v2[:,i]=np.power(abs(np.fft.fft(y[:,i])),2)
```

```python
vavg=np.zeros(512)
vavg=np.mean(v2, axis=1)
```
